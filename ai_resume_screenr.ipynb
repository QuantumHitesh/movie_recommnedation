```python
#@title Text formatting setup
#code is not necessary for colab.ai, but is useful in fomatting text chunks
import sys

class LineWrapper:
    def __init__(self, max_length=80):
        self.max_length = max_length
        self.current_line_length = 0

    def print(self, text_chunk):
        i = 0
        n = len(text_chunk)
        while i < n:
            start_index = i
            while i < n and text_chunk[i] not in ' \n': # Find end of word
                i += 1
            current_word = text_chunk[start_index:i]

            delimiter = ""
            if i < n: # If not end of chunk, we found a delimiter
                delimiter = text_chunk[i]
                i += 1 # Consume delimiter

            if current_word:
                needs_leading_space = (self.current_line_length > 0)

                # Case 1: Word itself is too long for a line (must be broken)
                if len(current_word) > self.max_length:
                    if needs_leading_space: # Newline if current line has content
                        sys.stdout.write('\n')
                        self.current_line_length = 0
                    for char_val in current_word: # Break the long word
                        if self.current_line_length >= self.max_length:
                            sys.stdout.write('\n')
                            self.current_line_length = 0
                        sys.stdout.write(char_val)
                        self.current_line_length += 1
                # Case 2: Word doesn't fit on current line (print on new line)
                elif self.current_line_length + (1 if needs_leading_space else 0) + len(current_word) > self.max_length:
                    sys.stdout.write('\n')
                    sys.stdout.write(current_word)
                    self.current_line_length = len(current_word)
                # Case 3: Word fits on current line
                else:
                    if needs_leading_space:
                        # Define punctuation that should not have a leading space
                        # when they form an entire "word" (token) following another word.
                        no_leading_space_punctuation = {
                            ",", ".", ";", ":", "!", "?",        # Standard sentence punctuation
                            ")", "]", "}",                     # Closing brackets
                            "'s", "'S", "'re", "'RE", "'ve", "'VE", # Common contractions
                            "'m", "'M", "'ll", "'LL", "'d", "'D",
                            "n't", "N'T",
                            "...", "…"                          # Ellipses
                        }
                        if current_word not in no_leading_space_punctuation:
                            sys.stdout.write(' ')
                            self.current_line_length += 1
                    sys.stdout.write(current_word)
                    self.current_line_length += len(current_word)

            if delimiter == '\n':
                sys.stdout.write('\n')
                self.current_line_length = 0
            elif delimiter == ' ':
                # If line is full and a space delimiter arrives, it implies a wrap.
                if self.current_line_length >= self.max_length:
                    sys.stdout.write('\n')
                    self.current_line_length = 0

        sys.stdout.flush()
```

## Develop resume evaluation logic

### Subtask:
Write code to evaluate each resume based on the defined screening criteria. This could involve using keyword matching, analyzing sentence structure, or employing more advanced natural language processing techniques.


**Reasoning**:
Define the `evaluate_resume` function to check for the presence of screening criteria in preprocessed resume tokens and then apply it to all preprocessed resumes.



```python
def evaluate_resume(resume_tokens, screening_criteria):
    """
    Evaluates a preprocessed resume based on defined screening criteria.

    Args:
        resume_tokens: A list of tokens representing the preprocessed resume.
        screening_criteria: A dictionary containing screening criteria
                            categorized by 'skills', 'experience', 'education',
                            and 'keywords'.

    Returns:
        A dictionary containing the evaluation results for the resume.
    """
    evaluation_results = {
        'skills': [],
        'experience': [],
        'education': [],
        'keywords': []
    }

    # Convert resume tokens to a single string for easier keyword/phrase matching
    resume_text = " ".join(resume_tokens)

    for category, items in screening_criteria.items():
        for item in items:
            # Simple check for presence of the item (case-insensitive)
            if item.lower() in resume_text:
                evaluation_results[category].append(item)

    return evaluation_results

# Apply the evaluate_resume function to each preprocessed resume
evaluation_results = [evaluate_resume(resume, screening_criteria) for resume in preprocessed_resumes]

# Print the evaluation results
print(evaluation_results)
```

```python
import streamlit as st
import string
import nltk
from nltk.corpus import stopwords
import io
from PyPDF2 import PdfReader # Using PyPDF2 for PDF text extraction

# Download NLTK resources once at the start
@st.cache_resource
def download_nltk_resources():
    nltk.download('punkt', quiet=True)
    nltk.download('stopwords', quiet=True)
    # punkt_tab is needed for word_tokenize with default language
    nltk.download('punkt_tab', quiet=True)

download_nltk_resources()

# Define screening criteria
screening_criteria = {
    'skills': [
        'Python', 'Machine Learning', 'Data Analysis', 'SQL', 'TensorFlow',
        'PyTorch', 'NLP', 'Computer Vision', 'Statistical Modeling', 'A/B Testing',
        'Spark', 'Hadoop', 'AWS', 'Azure', 'GCP', 'Docker', 'Kubernetes'
    ],
    'experience': [
        '5+ years in data science', 'Experience with large datasets',
        'Proven track record of deploying models', 'Experience in a tech company',
        'Experience building data pipelines', 'Experience with cloud platforms',
        'Experience with MLOps'
    ],
    'education': [
        "Master's degree in a quantitative field", 'PhD in Computer Science',
        'Master\'s degree in Data Science', 'PhD in Statistics'
    ],
    'keywords': [
        'Deep Learning', 'Artificial Intelligence', 'Data Mining', 'Big Data',
        'Predictive Modeling', 'Natural Language Processing', 'Computer Vision',
        'Machine Learning Engineer', 'Data Scientist', 'Data Engineer'
    ]
}


def preprocess_resume(resume_text):
    """
    Preprocesses a resume string by converting to lowercase, removing punctuation,
    removing stop words, and tokenizing.
    """
    # Convert to lowercase
    resume_text = resume_text.lower()

    # Remove punctuation
    resume_text = resume_text.translate(str.maketrans('', '', string.punctuation))

    # Tokenize
    tokens = nltk.word_tokenize(resume_text)

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word not in stop_words]

    return " ".join(filtered_tokens) # Return as a string for simplicity


def extract_text_from_pdf(pdf_file):
    """Extracts text from a PDF file."""
    text = ""
    try:
        reader = PdfReader(pdf_file)
        for page_num in range(len(reader.pages)):
            text += reader.pages[page_num].extract_text() or ""
    except Exception as e:
        st.error(f"Error reading PDF file: {e}")
        return None
    return text

def evaluate_resume(resume_text, screening_criteria):
    """
    Evaluates a preprocessed resume (as a string) based on defined screening criteria.

    Args:
        resume_text: A string representing the preprocessed resume.
        screening_criteria: A dictionary containing screening criteria
                            categorized by 'skills', 'experience', 'education',
                            and 'keywords'.

    Returns:
        A dictionary containing the evaluation results for the resume.
    """
    evaluation_results = {
        'skills': [],
        'experience': [],
        'education': [],
        'keywords': []
    }

    # Use the preprocessed text directly for simple string matching
    for category, items in screening_criteria.items():
        for item in items:
            # Simple check for presence of the item (case-insensitive)
            if item.lower() in resume_text:
                evaluation_results[category].append(item)

    return evaluation_results


def main():
    """Main function for the Streamlit Resume Screener application."""
    st.title("AI Resume Screener")
    st.write("Upload resumes and screen them based on predefined criteria.")

    uploaded_files = st.file_uploader(
        "Upload Resumes",
        type=['txt', 'pdf'],
        accept_multiple_files=True
    )

    preprocessed_resumes_data = [] # Store dicts with filename and preprocessed text

    if uploaded_files is not None:
        if len(uploaded_files) > 0:
            st.write("Processing Uploaded Files:")
            for file in uploaded_files:
                st.write(f"- {file.name}")

                raw_text = None
                if file.type == "text/plain":
                    # Read text file
                    stringio = io.StringIO(file.getvalue().decode("utf-8"))
                    raw_text = stringio.read()
                elif file.type == "application/pdf":
                    # Read PDF file
                    raw_text = extract_text_from_pdf(file)

                if raw_text:
                    # Preprocess the extracted text
                    preprocessed_text = preprocess_resume(raw_text)
                    preprocessed_resumes_data.append({"filename": file.name, "preprocessed_text": preprocessed_text})
                    # Display a snippet of the preprocessed text
                    st.write(f"  Preprocessed snippet: {preprocessed_text[:200]}...")

    # Evaluate preprocessed resumes
    evaluation_results = []
    if preprocessed_resumes_data:
        st.write("\nEvaluating Resumes:")
        for resume_data in preprocessed_resumes_data:
            filename = resume_data["filename"]
            preprocessed_text = resume_data["preprocessed_text"]
            st.write(f"  Evaluating {filename}...")
            results = evaluate_resume(preprocessed_text, screening_criteria)
            evaluation_results.append({"filename": filename, "evaluation": results})

        # Placeholder for displaying results
        print("Resume evaluation completed. Results stored in evaluation_results.")


if __name__ == "__main__":
    main()
```
